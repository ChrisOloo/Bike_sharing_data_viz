```python
# importing the libraries to use

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests

# allow matplotlip to show interactive images within the notebook
%matplotlib inline
```

```python
# the public api url
url = "https://gbfs.baywheels.com/gbfs/es/station_status.json"

# getting the raw data with requests
raw_data = requests.get(url)

# instatiating the raw data collected in json
data = raw_data.json()

# having a general look of the raw data collected in json
data
```

```python
# identifying the keys in the data
data_keys = data.keys()
data_keys
```

```python
# getting the list of the station data
stations_data = data["data"]["stations"]
stations_data
```

```python
# converting the station data to pandas dataframe
df = pd.DataFrame(stations_data)
df
```

```python
# have a general description of the data's dataframe
df.describe()
```

```python
# get the shape of the data (rows, cols)
df.shape
```

```python
# get general info of the data
df.info()
```

```python
# get a look of the head of the data
df.head()
```

```python
# sorting the data based on the number of bikes available in the station from the bottom
df_sorted = df.sort_values(by="num_bikes_available")

# display the details of the bottom 20 stations with bikes available
bottom_20 = df_sorted.head(20)
bottom_20
```

```python
# setting the figsize for the plot
plt.figure(figsize=(12, 12))

# Change default settings
sns.set(context='talk')

# bar plot using seaborn for the stations with least no. of bikes
sns.barplot(data=bottom_20, y=bottom_20["station_id"], x=bottom_20["num_bikes_available"]).set(xlim=(0, 30));
plt.xlabel("Number of bikes availabe")
plt.ylabel("Stations IDs")
plt.title("Stations with the Least number of Bikes Available")

# saving the visualisation
plt.savefig("bottom_20.png")
```

```python
# display the details of the top 20 stations with bikes available
top_20 = df_sorted.tail(20)
top_20
```

```python
# setting the figsize for the plot
plt.figure(figsize=(12, 12))

# Change default settings
sns.set(style='whitegrid', context='talk')

# bar plot using seaborn for the stations with least no. of bikes
sns.barplot(data=top_20, y=top_20["station_id"], x=top_20["num_bikes_available"]);
plt.xlabel("Number of bikes availabe")
plt.ylabel("Stations IDs")
plt.title("Stations with the Most number of Bikes Available")

# saving the visualisation
plt.savefig("top_20.png")
```

```python
# categorizing the number of stations based on the number of bikes they have
df_bikes = df['num_bikes_available'].value_counts()

# converting the pandas series to pandas dataframe
df_bikes = df_bikes.to_frame()

# renaming for accuracy
df_bikes.rename(columns={'num_bikes_available': 'num_of_stations'}, inplace=True)

# creating a new column of num_bikes_availabe 
df_bikes['num_bikes_available'] = df_bikes.index

# dropping the index which was based on the df_g1['num_bikes_availabe']
df_bikes.reset_index(drop=True)
```

```python
# the shape of the new dataframe categorizing the no. of bikes to no. of stations with those bikes
df_bikes.shape
```

```python
# setting the figsize for the plot
plt.figure(figsize=(20, 12))

# Change default settings
sns.set(style='whitegrid', context='talk')

# bar plot using seaborn for the stations with least no. of bikes
sns.barplot(data=df_bikes, y=df_bikes["num_of_stations"], x=df_bikes["num_bikes_available"]);
plt.xlabel("Number of bikes availabe")
plt.ylabel("Number of Stations")
plt.title("Number of Bikes Available per number of Stations")

# saving the visualisation
plt.savefig('overal_load.png')
```
